```python
import pandas as pd
import tensorflow as tf
import keras
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.info()
```

```python
df1 = df.drop(labels=["RowNumber","CustomerId","Surname"],axis=1)
df1.head()
```

```python
df_gender = pd.get_dummies(df['Gender'],drop_first=True)
df_Geo = pd.get_dummies(df['Geography'],drop_first=True)
```

```python
df1.drop(labels=['Geography','Gender'],axis=1, inplace=True)
df1.head()
```

```python
df1 = pd.concat([df_gender,df_Geo,df1],axis=1)
df1.head() 
```

```python
y = df1["Exited"]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df1.drop(['Exited'],1)
X.head()
X.columns
```

```python
Scaler = StandardScaler()
```

```python
X = pd.DataFrame(Scaler.fit_transform(X),columns= X.columns)
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
```

```python
NN_model = Sequential()

#Adding the input layer
NN_model.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=11))

#Adding second layer(hidden layer 1)
NN_model.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))

#Adding Output layer
NN_model.add(Dense(units=1,kernel_initializer='he_uniform',activation='sigmoid'))

#Compile ANN
NN_model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

NN_model.fit(X_train,y_train,batch_size=10,
    epochs=50,validation_split=0.33)
```

```python
NN_model.summary()
```

```python
y_pred = NN_model.predict_classes(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
score = accuracy_score(y_test,y_pred)
```

```python
score
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```
